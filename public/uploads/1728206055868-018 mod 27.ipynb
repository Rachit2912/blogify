````python
### Q1. What is Flask Framework? What are the advantages of Flask Framework?
'''
**Flask** is a lightweight and flexible web framework for Python, used for building web applications. It follows the WSGI standard and the "micro-framework" approach.

**Advantages**:
- **Lightweight**: Minimal structure, allowing flexibility.
- **Simple to use**: Easy setup with fewer configurations.
- **Extensible**: Supports integration of third-party libraries.
- **Modular**: Allows building modular applications with extensions.

---
'''

### Q2. Simple Flask Application to Display "Hello World!!"
```python
from flask import Flask
app = Flask(__name__)

@app.route('/')
def hello_world():
    return 'Hello World!!'

if __name__ == '__main__':
    app.run()
```



---

### Q3. What is App Routing in Flask? Why do we use App Routes?
'''
**App routing** in Flask maps a URL to a specific function that handles the request. It defines the URL structure of the web app.

**Why use App Routes?**
- Helps in organizing URLs for different pages.
- Allows linking functions with URLs for specific HTTP requests.

---
'''

### Q4. Flask Routes for Welcome and Company Details
from flask import Flask
app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def company_details():
    return '''Company Name: ABC Corporation<br>
              Location: India<br>
              Contact Detail: 999-999-9999'''

if __name__ == '__main__':
    app.run()


### Q5. URL Building with `url_for()`
'''
In Flask, the `url_for()` function dynamically builds URLs for specific routes.


**Example**:
'''

from flask import Flask, url_for
app = Flask(__name__)

@app.route('/')
def index():
    return 'Home Page'

@app.route('/profile/<username>')
def profile(username):
    return f'Profile page of {username}'

with app.test_request_context():
    print(url_for('index'))
    print(url_for('profile', username='JohnDoe'))

````
